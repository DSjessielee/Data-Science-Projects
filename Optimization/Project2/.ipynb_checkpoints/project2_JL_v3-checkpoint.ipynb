```python
import numpy as np
import pandas as pd
import gurobipy as gp
```

# Load and Prepare Data

```python
stk_19=pd.read_csv('stocks2019.csv')
stk_20=pd.read_csv('stocks2020.csv')
```

```python
stk_rt_19=pd.DataFrame(np.zeros((len(stk_19)-1,len(stk_19.columns))))
stk_rt_20=pd.DataFrame(np.zeros((len(stk_20)-1,len(stk_20.columns))))
stk_rt_19.columns=stk_19.columns
stk_rt_20.columns=stk_20.columns
```

```python
stk_19
```

```python
# Convert from price to return for 2019 data
for i in range(stk_19.shape[0]-1):
    for j in range(1,stk_19.shape[1],1):
        stk_rt_19.iloc[i,j]=stk_19.iloc[i+1,j]/stk_19.iloc[i,j]-1

# Convert from price to return for 2020 data
for i in range(stk_20.shape[0]-1):
    for j in range(1,stk_19.shape[1],1):
        stk_rt_20.iloc[i,j]=stk_20.iloc[i+1,j]/stk_20.iloc[i,j]-1
        
```

```python
# Index Tickers
ticker=stk_20.columns[2:]

# change matrix
rt19=np.array(stk_rt_19.iloc[:,2:])
rt20=np.array(stk_rt_20.iloc[:,2:])
# return correlation matrix
cor19=np.corrcoef(rt19, rowvar=False) 
cor20=np.corrcoef(rt20, rowvar=False) 

idx_rt19 = np.array(stk_rt_19.iloc[:, 1:2])
idx_rt20 = np.array(stk_rt_20.iloc[:, 1:2])
```

# Start with 5 stocks

```python
m=5
n=cor19.shape[1]
```

```python
A=np.zeros((n**2+n+1,n**2+n))
sense=np.array(['']*A.shape[0])
b=np.zeros((A.shape[0])) 
```

Constraint part I: $\sum \limits _{i=0} ^{100} X_{ij} $ =1   ..........100 constraints

```python
# Constraint: 

for i in range(n):
    for j in range(n):
        A[i,n*i+j]=1
    sense[i]='='
    b[i]=1
```

Constraint part II: $X_{ij}<Y_{j} $    ...........100*100 constraints

```python
for i in range(n):
    for j in range(n):
        A[i * n + j + n, i * n + j]=1
        A[i * n + j + n, j + n**2]=-1
        sense[i * n + j + n]='<'
        b[i * n + j + n]=0
```

Constraint part III: $\sum \limits _{j=0} ^{n=100}y_{j}<m $    ...........1 constraints

```python
#Weight cannot be larger than 0 if it's not selected
for j in range(n):
    A[n**2 + n, n**2 + j ] = 1
sense[n**2 + n] = '='
b[n**2 + n] = m
```

# Objective Function

```python
obj=np.hstack((np.ndarray.flatten(cor19),np.array([0]*n)))
```

# Variable type

```python
var_type=['B']*(n**2+n)
```

# Model

```python
StkMod = gp.Model()
StkMod_x = StkMod.addMVar(len(obj),vtype=var_type) # vtype can be: 'C' or 'I' or 'B'
StkMod_con = StkMod.addMConstrs(A, StkMod_x, sense, b)
StkMod.setMObjective(None,obj,0,sense=gp.GRB.MAXIMIZE)

StkMod.Params.OutputFlag = 0 # tell gurobi to shut up!!
StkMod.optimize()
```

```python
# Max coefficient
StkMod.objVal
```

```python
import math
port_idx=[]
for i in range(n**2,len(obj),1):
    if StkMod_x.x[i]==1:
        port_idx.append(i-n**2)
```

# Optimize weight matrix for 5 chosen stocks

```python
#250 return data per stock
t=rt19.shape[0]
```

```python
# Initialize Constraint Matrix
A=np.zeros((t * 2 + 1 , t + m)) 
sense=np.array(['']*A.shape[0])
b=np.zeros((A.shape[0])) 
```

```python

for i in range(t):
    for j in range(m):
        #A[:5] are the weight of chosen stocks
        A[i * 2, j] = rt19[i, port_idx[j]]
        A[i * 2 + 1, j] = rt19[i, port_idx[j]]
    # A[6:6+250] --> y are the difference between weighted gain and index gain 
    # w.x - qt - y <= 0
    A[i * 2, i + m] = -1
    sense[i * 2] = '<'
    b[i * 2] = idx_rt19[i]
    #w.x - qt + y >= 0
    A[i * 2 + 1, i + m] = 1
    sense[i * 2 + 1] = '>'
    b[i * 2 + 1] = idx_rt19[i]

#Sum of weight = 1
for i in range(m):
    A[t*2, i] = 1
b[t*2] = 1
sense[t*2] = '='

# Objective to minimize

obj = np.hstack((np.zeros((m)),np.array([1]*t)))

# Variable type

var_type=['C']*A.shape[1]
```

```python
# Optimize
StkMod = gp.Model()
StkMod_x = StkMod.addMVar(len(obj),vtype=var_type) # vtype can be: 'C' or 'I' or 'B'
StkMod_con = StkMod.addMConstrs(A, StkMod_x, sense, b)
StkMod.setMObjective(None,obj,0,sense=gp.GRB.MINIMIZE)

StkMod.Params.OutputFlag = 0 # tell gurobi to shut up!!
StkMod.optimize()
```

```python
diff19=StkMod.objVal
round(diff19,4)
```

# Repeat with m =  10, 20, …, 90, 100

```python
m=np.linspace(10,100,10)
stk_chosen=dict(zip(m,['']*len(m)))
wt_stk_chosen=dict(zip(m,['']*len(m)))
```

```python

for m in np.linspace(10,100,10):
    
    n=cor19.shape[1]
    A=np.zeros((n**2+n+1,n**2+n))
    sense=np.array(['']*A.shape[0])
    b=np.zeros((A.shape[0])) 
    for i in range(n):
        for j in range(n):
            A[i,n*i+j]=1
        sense[i]='='
        b[i]=1

    for i in range(n):
        for j in range(n):
            A[i * n + j + n, i * n + j]=1
            A[i * n + j + n, j + n**2]=-1
            sense[i * n + j + n]='<'
            b[i * n + j + n]=0
    
    for j in range(n):
        A[n**2 + n, n**2 + j ] = 1
        sense[n**2 + n] = '='
        b[n**2 + n] = m
    
    for j in range(n):
        A[n**2 + n, n**2 + j ] = 1
        sense[n**2 + n] = '='
        b[n**2 + n] = m
    
    obj=np.hstack((np.ndarray.flatten(cor19),np.array([0]*n)))

    var_type=['B']*(n**2+n)
    
    StkMod = gp.Model()
    StkMod_x = StkMod.addMVar(len(obj),vtype=var_type) # vtype can be: 'C' or 'I' or 'B'
    StkMod_con = StkMod.addMConstrs(A, StkMod_x, sense, b)
    StkMod.setMObjective(None,obj,0,sense=gp.GRB.MAXIMIZE)

    StkMod.Params.OutputFlag = 0 # tell gurobi to shut up!!
    StkMod.optimize()
    
    port_idx=[]
    for i in range(n**2,len(obj),1):
        if StkMod_x.x[i]==1:
            port_idx.append(i-n**2)
    
        stk_chosen[m]=port_idx
```

# Optimize weight matrix for different m

```python
stk_wt_all_name=[]
stk_wt_all_idx=[]

for stk in stk_chosen.values():
    m=len(stk)
    t=rt19.shape[0]
    A=np.zeros((t * 2 + 1 , t + m)) 
    sense=np.array(['']*A.shape[0])
    b=np.zeros((A.shape[0])) 
    for i in range(t):
        for j in range(len(stk)):
            #A[:5] are the weight of chosen stocks
            A[i * 2, j] = rt19[i, stk[j]]
            A[i * 2 + 1, j] = rt19[i, stk[j]]
        # A[6:6+250] --> y are the difference between weighted gain and index gain 
        # w.x - qt - y <= 0
        A[i * 2, i + m] = -1
        sense[i * 2] = '<'
        b[i * 2] = idx_rt19[i]
        #w.x - qt + y >= 0
        A[i * 2 + 1, i + m] = 1
        sense[i * 2 + 1] = '>'
        b[i * 2 + 1] = idx_rt19[i]

    #Sum of weight = 1
    for i in range(m):
        A[t*2, i] = 1
        b[t*2] = 1
        sense[t*2] = '='

    # Objective to minimize

    obj = np.hstack((np.zeros((m)),np.array([1]*t)))

    # Variable type

    var_type=['C']*A.shape[1]
    
    # Optimize
    StkMod = gp.Model()
    StkMod_x = StkMod.addMVar(len(obj),vtype=var_type) # vtype can be: 'C' or 'I' or 'B'
    StkMod_con = StkMod.addMConstrs(A, StkMod_x, sense, b)
    StkMod.setMObjective(None,obj,0,sense=gp.GRB.MINIMIZE)

    StkMod.Params.OutputFlag = 0 # tell gurobi to shut up!!
    StkMod.optimize()
    
    stk_wt_name=dict(zip([ticker[i] for i in stk], StkMod.X[:m]))
    stk_wt_idx=dict(zip([i for i in stk],StkMod.X[:m]))
    stk_wt_all_name.append(stk_wt_name)
    stk_wt_all_idx.append(stk_wt_idx)
```

# 4) Another way to select stocks by minimizing difference

```python
#Period of return ~250
t=rt19.shape[0]
#of tickers ~100
n=rt19.shape[1]
#of tickers in portfolio
m=list(np.linspace(10,100,10))
#Big M=1
M=1
```

```python
port_comp2=pd.DataFrame(columns=['difference for 2019','difference for 2020'],
                       index=['m=10','m=20','m=30','m=40','m=50','m=60','m=70','m=80','m=90','m=100'])
```

```python
b[100:350]=idx_rt19.T
b[350:600]=idx_rt19.T
```

```python
#For all Xi, Xi - Yi <= 0.

diff=[]
weight=[]
for m in np.linspace(10,100,10):
            
            # first 100 var are weights, second 200 are if a stock is selected (binary) for portfolio
            # third batch 250 are the difference between daily return of portfolio and index funds
            
            #obj[0X200+1X250]
            obj=np.array([0]*n*2+[1]*t)       # 450 variables
            
            #A.shape --> 602X450
            A=np.zeros((n+t*2+2,2*n+t))       # 602 constraints
            
            sense=['']*A.shape[0]  
            
            b=np.zeros((A.shape[0]))

            row = 0
            # weight has to be 0 if the stk is not selected - 100 constraints
            for i in range(n):

                A[row,i]=1
                A[row,i+n]=-M
                sense[row]='<'
                row+=1
                
            # difference between weighted average of portfolio and index for each day (X-diff<index) 250 constrains
            for i in range(t):
                
                A[row,:n]=rt19[i]
                A[row,2*n+i]=-1
                sense[row]='<'
                b[row]=idx_rt19[i][0]
                row+=1

            # difference between weighted average of portfolio and index for each day (X+diff>index) 250 constrains
            for i in range(t):
                
                A[row,:n]=rt19[i]
                A[row,2*n+i]=1
                sense[row]='>'
                b[row]=idx_rt19[i][0]
                row+=1
                
            # all weights sum up to 1 ..........1 constraint
            A[row,:n]=1
            sense[n+2*t]='='
            b[n+2*t]=1
            row+=1
            
            # total of m selected stocks  ...........1 constraint
            A[row,n:2*n]=1
            sense[row]='='
            b[row]=m
            

            v_type=['C']*100 + ['B'] * 100 + ['C'] * 250

            StkMod = gp.Model()
            StkMod_x = StkMod.addMVar(len(obj),vtype=v_type) # 'C' or 'I' or 'B'
            StkMod_con = StkMod.addMConstrs(A, StkMod_x, sense, b)
            StkMod.setMObjective(None,obj,0,sense=gp.GRB.MINIMIZE)

            StkMod.Params.OutputFlag = 0 # tell gurobi to shut up!!
            StkMod.Params.TimeLimit = 3600
            StkMod.optimize()
            
            
            diff.append(StkMod.objVal)
            weight.append(StkMod_x.X[:n])

diff19=pd.DataFrame(diff,columns=['difference in 2019'])
wt19=pd.DataFrame(weight)
diff19.to_csv('diff19.csv',index=False)
wt19.to_csv('wt19.csv',index=False)

port_comp2['difference for 2019']=diff
for i in range(len(weight)):
    port_comp2['difference for 2020'][i]=sum(abs(np.sum(weight[i]*rt20,axis=0)-idx_rt20[i]))
port_comp2
```
