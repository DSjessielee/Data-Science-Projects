### **Question 1**
1	Rock, Paper, Scissors (20 points)

This is a popular game (see here), and you must create a computer player.
Game setup. Each game round consists two turns, the first by the computer and the second by a human. The computer continues playing rounds until the human chooses to quit.

•	The computer chooses one of Rock, Paper, and Scissors, but keeps its choice secret.

•	The computer asks for the human’s input.

•	The human chooses one of Rock, Paper, and Scissors, or Quit.

•	Unless the human quits, the computer figures out the result of the game, as follows:

        –	Rock smashes Scissors, so Rock beats Scissors.

        –	Scissors can cut up paper, so Scissors beat Paper. 
        
        –    Paper covers Rock, so Paper beats Rock.

If both players chose the same, it is a draw. The computer reports the result of this round.


•	If the human chooses to quit, the computer reports:

–	the number of games played, and – the number of times the human won.

**Computer’s brains.** The computer must be able to exploit some human biases. If the human has played Rock most often, the computer should assume that he or she will play Rock in the next round, so the computer should play Paper. If the human has played Rock and Paper equally often, and Scissors less often, the computer should assume that the human is going to play either Rock or Paper (both equally likely) in the next round. (What should the computer play?)

Hence, **your program should remember how many Rock, Paper, or Scissors were played by the human.** Note that we don’t need to remember the order in which the human chooses these; the total counts so far for each choice will be enough.
Gotchas.

•	User input: How you want to receive the user’s input is up to you, but you must check the user’s input to make sure it is valid (you can assume that the user input is of the correct type). If it isn’t, request the user for input again.



```python
import random
```

```python
# Take player input of choice from Rock, Scissor and Paper

def player_input():
    '''
    Take player input of choice from Rock, Scissor and Paper
    '''
    player_choice=''
    while player_choice not in {'R','S','P'}:
        player_choice = input('Player please make your choice - Rock(R), Scissors(S) or Paper (P): ')
        player_choice = player_choice.upper()
    return (player_choice)

#Decise who wins

def who_win(player_choice,computer_choice):
    '''
    Take player choice and computer choice to decise who wins
    '''
    if player_choice=='R':
        if computer_choice=='S':
            print('Rock smashes scissors! Player wins!')
            return ('player win')
        elif computer_choice=='P':
            print('Paper covers rock. Computer wins!')
            return ('computer win')
        else:
            print('You both played rock. Its a draw!'.format(computer_choice))
            return ('draw')
    elif  player_choice =='P':
        if computer_choice=='R':
            print('Paper covers rock. Player wins!')
            return ('player win')
        elif computer_choice=='S':
            print('Scissors cut paper. Computer wins!')
            return ('computer win')
        else:
            print('You both played paper. Its a draw!')
            return ('draw')
    else:
        if computer_choice=='P':
            print('Scissors cut paper. Player wins!')
            return ('player win')
        elif computer_choice=='R':
            print('Rock smashes scissors! Computer wins!')
            return ('computer win')
        else:
            print('You both played scissors. Its a draw!')
            return ('draw')
    
```

```python
#Computer made 'trained' decision

from numpy.random import choice
def computer_learning_choice(player_choice_rec):
    '''
    Computer will always choose the R, S, or P that beats what the human player picked most frequently.
    '''
    r_count=player_choice_rec.count('R')
    s_count=player_choice_rec.count('S')
    p_count=player_choice_rec.count('P')
    
    count_rec=[r_count,s_count,p_count]

#if the probability for each choice is equal, choose randomly    
    if r_count==s_count==p_count:
        computer_choice = choice(['R','S','P'])
#if not equal,then there could be the probability of 2 of them are equal, or all of them are different...   
    elif r_count==max(count_rec):
        if r_count==s_count:
            computer_choice='R'
        else:
            computer_choice='P'
    elif s_count==max(count_rec):
        if s_count==p_count:
            computer_choice='S'
        else:
            computer_choice='R'
    else:
        if p_count==r_count:
            computer_choice='P'
        else:
            computer_choice='S'
            
    return(computer_choice)
        

### Alternative algrithm to make 'weighted' ramdon choice based on player choices###
# If computer will consider all entries and form a strategy not to loose, here is an alternative
#     r_count=player_choice_rec.count('R')/len(player_choice_rec)
#     s_count=player_choice_rec.count('S')/len(player_choice_rec)
#     p_count=player_choice_rec.count('P')/len(player_choice_rec)
#     r_weight=(r_count+s_count)/2
#     s_weight=(s_count+p_count)/2
#     p_weight=(p_count+r_count)/2
#     weight=[r_weight,s_weight,p_weight]
#     rsp=['R','S','P']
#     computer_choice = choice(rsp,p=weight)
```

```python
def game():
    player_choice_rec=['R','S','P']
    computer_win_ct=0
    player_win_ct=0
    player_choice=['']
    while True:
        player_choice=player_input()
        computer_choice=computer_learning_choice(player_choice_rec)
        result = who_win(player_choice,computer_choice)
        
        if result=='player win':
            player_win_ct+=1
        if result=='computer win':
            computer_win_ct+=1
        player_choice_rec.append(player_choice)
        game_on=''
        while game_on.upper() not in {'Y','N'}:
            game_on=input('Do you want to play again? Y or N ')
        if game_on.upper()=='N':
            break
    your_win_rate=player_win_ct/(player_win_ct+computer_win_ct)*100
    
    print('You play {} times!\n'.format(len(player_choice_rec)-3))
    print('You win {} times!\n'.format(player_win_ct))
    print('Your win rate is {}%! Good job!'.format(round(your_win_rate,2)))
```

```python
game()
```
