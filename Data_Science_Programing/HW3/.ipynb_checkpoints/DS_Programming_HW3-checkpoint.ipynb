```python
# how is rating counts correlated with accuracy?

import statsmodels.api as sm
from patsy import dmatrices

all_turk=rater_goodness.copy()
all_turk.columns=['rating_count','correctness','odds']

#prepare the variable and target
y, X = dmatrices('correctness ~ rating_count', data=all_turk, return_type='dataframe')

#try a linear regression using odds as the only variable
model = sm.OLS(y, X)       # Set up the model
result = model.fit()       # Fit model (find the intercept and slopes)
print(result.summary())
plt.scatter(x=all_turk['rating_count'],y=all_turk['correctness'])
```

```python
#try 'odds' for the more frequence raters?
freq_turk=all_turk[all_turk['rating_count']>=20]

#prepare the variable and target
y, X = dmatrices('correctness ~ rating_count', data=freq_turk, return_type='dataframe')

#try a linear regression using odds as the only variable
model = sm.OLS(y, X)       # Set up the model
result = model.fit()       # Fit model (find the intercept and slopes)
print(result.summary())
plt.scatter(x=freq_turk['rating_count'],y=freq_turk['correctness'])
```


**No obvious correlation between rater counts and correctness ratio

**The correlation coefficient and R2 are both very low. P value is high and we cannot reject null hypothsis that they are not correlated.

**If we do the same exercise for only the frequent raters (rater count >=20), still no correlation observed.

easier to have ~100% correctness with low number of ratings. Howeever, turks with most votes have ~70% correctness. 

# Question 7

```python
reliable_raters = rater_goodness.sort_values(by="Number of Ratings", ascending=False)
#Creating a mask for when the number of ratings is greater than 75% of all turks
reliable_raters_top_25 = reliable_raters[reliable_raters["Number of Ratings"] > np.quantile(reliable_raters["Number of Ratings"], .75)]
reliable_raters_top_25
```

```python
reliable_raters_top_25 = reliable_raters_top_25.merge(labels_unknown,
                                        left_on="turk",
                                        right_on="turk")
reliable_raters_top_25
```

```python
#Creating heirarchical index on URL and Category and calculating the product of odds for this grouping
reliable_raters_odds = reliable_raters_top_25.groupby(['url','category'])['odds'].agg(['prod'])
reliable_raters_odds = reliable_raters_odds.rename(columns = {'prod':'Overall Odds'})
reliable_raters_odds
```

# Question 8

```python
#Picking the category with the best odds for each URL
result_75 = reliable_raters_odds.loc[reliable_raters_odds.groupby("url")['Overall Odds'].idxmax()]
result_75 = result_75.rename(columns = {"Overall Odds":"Top Odds"})
result_75 = result_75.reset_index().set_index('url')
result_75
```

# Question 9

```python
#repeating above process for turks that are in the top 75%
reliable_raters_top_75 = reliable_raters[reliable_raters["Number of Ratings"] > np.quantile(reliable_raters["Number of Ratings"], .25)]
reliable_raters_top_75
```

```python
reliable_raters_top_75 = reliable_raters_top_75.merge(labels_unknown,
                                        left_on="turk",
                                        right_on="turk")
reliable_raters_top_75_odds = reliable_raters_top_75.groupby(['url','category'])['odds'].agg(['prod'])
result_25 = reliable_raters_top_75_odds.loc[reliable_raters_top_75_odds.groupby('url')['prod'].idxmax()]
result_25 = result_25.rename(columns = {"prod":"Top Odds"})
result_25 = result_25.reset_index().set_index('url')
result_25
```

```python
#resetting indices to enable crosstab
result_25_df = result_25.reset_index()
result_75_df = result_75.reset_index()
result_25_df
```

```python
#merging tables so we can get common URLS
merge = result_25_df.merge(result_75_df,
                        left_on='url',
                        right_on='url')
merge
```

```python
x_tab=pd.crosstab(merge['category_y'],merge['category_x'],rownames=["Result_75"],colnames=["Result_25"])
x_tab
```

* Most errors are in the [G,P] position. 586 urls have been rated G by the top 25% turks but P by the top 75%
