```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

```python
golddf = pd.read_csv("gold.txt",sep='\t',names=['url','category'])
labelsdf = pd.read_csv("labels.txt",sep='\t',names=['turk','url','category'])
```

```python
labelsdf.head()
golddf.head()
```

```python
labels_on_gold = pd.DataFrame(columns = ['turk', 'url', 'category'])
labels_unknown = pd.DataFrame(columns = ['turk', 'url', 'category'])
```

```python
labels_on_gold = labelsdf[labelsdf['url'].isin(golddf["url"])]
labels_unknown = labelsdf[~labelsdf['url'].isin(golddf["url"])]
```

```python
print(len(labels_on_gold), len(labels_unknown))
```

```python
category_match = labels_on_gold.merge(golddf, on="url", how="inner", suffixes=['_labels','_gold'])
category_match
```

```python
rater_goodness = category_match.groupby('turk')['category_check'].agg(['count','mean'])
rater_goodness = rater_goodness.rename(columns = {'count':'Number of Ratings','mean':'Average Correctness'})
rater_goodness
```

```python
rater_goodness['odds'] = rater_goodness['Average Correctness']/(1.001 - rater_goodness['Average Correctness'])
rater_goodness
```

easier to have ~100% correctness with low number of ratings. Howeever, turks with most votes have ~70% correctness. 

```python
reliable_raters = rater_goodness.sort_values(by="Number of Ratings", ascending=False)
reliable_raters_top_25 =reliable_raters[reliable_raters["Number of Ratings"] > np.quantile(reliable_raters["Number of Ratings"], .75)]
reliable_raters_top_25























```

```python
reliable_raters_top_25 = reliable_raters_top_25.merge(labels_unknown,
                                       on="turk",
                                       how="inner")

reliable_raters_top_25
```

```python
reliable_raters_odds = reliable_raters_top_25.groupby(['url','category'])['odds'].agg(['prod'])
reliable_raters_odds = reliable_raters_odds.rename(columns = {'prod':'Overall Odds'})
reliable_raters_odds
```
