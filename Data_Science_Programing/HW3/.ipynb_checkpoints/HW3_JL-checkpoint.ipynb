# Introduction
Suppose we have a bunch of URLs and we want to know their adult-rating
(i.e., is the url P, or G, or X, or R). This task is difficult for computers,
but easy for humans, and this has led to the growth of crowdsourcing: get
a bunch of humans to give ratings to urls, but use automated techniques to
figure out how much to trust each person’s ratings. We are going to use the data from a 
paper by Ipeirotis et al. available here. This details an experiment run on Amazon’s Mechanical 
Turk crowdsourcing system. They ask a bunch of raters (called “turks”) to rate several urls, but 
they already know the answers (the true categories) for a few urls,
called the “gold set”. The ratings of the turks on the gold set thus allows us
to judge their accuracy

# [Q2 10 points] Split into two DataFrames
Split the labels DataFrame into two: labels on gold and labels unknown,
the former containing all rows where the url is present in the gold set, and
the latter one contains all remaining rows of labels.

```python
gold_url=list(gold['url'])
def match(x):
    if x in gold_url:
        return True
    else:
        return False
    
label['match']=label['url'].map(match)

match_gold=label[label['match']==True]
no_match_gold=label[label['match']==False]
```

# [Q3 10 points] Compute accuracies of turks

Create a rater goodness DataFrame that is indexed by turk, and has two
columns: the number of ratings, and the average correctness of ratings for
each turk (both on gold set urls).

```python
#merge the match set with gold set
temp=match_gold.merge(gold,
                         left_on='url',
                         right_on='url',
                         suffixes=['_m','_g']
                         )

compare=temp[['turk','url','category_m','category_g']]
compare.shape
```

```python
#judge correctness of ratings

compare['correctness']=compare['category_m']==compare['category_g']
compare[100:120]
```

```python
#findout who are correct
df=compare.groupby(['turk','correctness'])[['correctness']].count().unstack('correctness')
df.fillna(0,inplace=True)

#calculate accuracy of turks
df['rater_count']=df[('correctness', False)]+df[('correctness',True)]
df['rater_correct_ratio']=df[('correctness',True)]/df['rater_count']

rater_correctness=df[['rater_count','rater_correct_ratio']]

print(rater_correctness.columns)
rater_correctness
```

# [Q7 13 points] Overall predicted odds
Consider each url u that is not in the gold set, and each category c. For the
pair (u, c), calculate the product of odds of all turks who (a) rated url u as
category c, and (b) have rated more gold set urls than 75% of all turks who
rated at least one gold-set url.

For example, if you find that there are 269 turks who rated at least one
gold-set url, you want to select only the turks who have rated more gold-set
urls than 75% of these 269 turks. We can think of these as our “reliable”
turks. Now, our belief that url u belongs to category c depends on how many
“reliable” turks rated u as c; specifically, our belief is based on the product
of their reliability scores (i.e., their odds).

We shall call such products of odds the overall odds henceforth.

```python
#find the top 25% raters

mask=rater_correctness['rater_count']>rater_correctness['rater_count'].quantile(0.75)
rater_correctness['turk']=rater_correctness.index

top_25_rater=list(rater_correctness[mask]['turk'])
print('There are {} total gold raters. \n'.format(len(rater_correctness)))
print('There are {} top 25% raters. \n'.format(len(top_25_rater)))
```

```python
no_match_gold
```

```python
#creat a column to idenify who is in top 25% rater group

def is_top_rater(x):
    if x in top_25_rater:
        return True
    else:
        return False

test=no_match_gold[['url','category','turk']]
test['top_25%_rater']=test['turk'].map(is_top_rater)
```

```python
#calculate the reliability score of each url rating category

test_score=test.groupby(['url','category','top_25%_rater'])['turk'].count().unstack('top_25%_rater').fillna(0)

test_score['reliable_rate_ratio']=test_score[True]/(test_score[False]+test_score[True])
test_score
```

```python
#find out ratio of reliable raters who rate the url to certain category
test_score2=test[test['top_25%_rater']==True].groupby(['url'])['turk'].count()
pd.DataFrame(test_score2)


test_set=test_score.merge(test_score2,
                left_index=True,
                right_index=True,
                )
test_set['rating_cate_ratio']=test_set[True]/test_set['turk']

test_set['overall_odds']=test_set['reliable_rate_ratio']*test_set['rating_cate_ratio']

test_set[20:40]
```

```python
#creat a dataframe with the Overall predicted odds

final_ratio=test_set['overall_odds']
final_ratio=pd.DataFrame(final_ratio)
```

# [Q8 13 points] Predicted categories
Create a DataFrame (called result 75, whose index is URLs not in the gold
set, and with two columns called top category and top odds. The top
