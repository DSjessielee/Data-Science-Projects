```python
vac_h_cln=vac_h[['State Name', '% Estimated Hesitant or Unsure - May 26 - June 7, 2021',
                 '% Estimated Hesitant -  May 26 - June 7, 2021',
                '% Estimated Strongly Hesitant -  May 26 - June 7, 2021']]
vac_h_cln.columns=['state','low','med','high']
vac_h_cln['tot_hesi']=vac_h_cln['low']+vac_h_cln['med']+vac_h_cln['high']
vac_h_cln.head()
```

# Merged Dataset for Modeling
**Vaccination\\vac_status

**Vaccine hesitation\\vac_h_cln

**Vaccination coverage concern\\vac_c_cln and vac_c_sub

**Political\\policical_cln

**Religion\\relig_cln

```python
merg=vac_status.merge(vac_c_cln,
              left_on='state',
              right_on='State')
merg=merg.merge(vac_h_cln,
              left_on='state',
                right_on='state')
merg=merg.merge(relig_cln,
               on='state')
merg=merg.merge(political_cln,
               on='state')
merg.sort_values(by='Administered_Dose1_Pop_Pct')[:10]['party']
merg.rename(columns={'Administered_Dose1_Pop_Pct': 'vac_rate'}, inplace=True)
```

```python
mod_inv = pd.get_dummies(merg, columns=['party'])
mod_inv[20:40]
```

```python
'''
This is to test the optimal max depth
@ n estimator=10

'''

min_depth=1
max_depth=100
d_step=1
rmse_depth=[]

for i in range(min_depth,max_depth,d_step):
    bty_mod=RandomForestRegressor(n_estimators=10, max_depth=i, max_leaf_nodes=500, min_samples_split=2, max_features='sqrt', random_state=42)
    bty_mod.fit(X_train,y_train)
    
    y_pred=bty_mod.predict(X_test)
    rmse_depth.append(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

time2=time.time()
time_diff=time2-time1

print('total time {}'.format(time_diff))
```

```python
'''
error table for 

max depth evaluation

'''

n=list(range(min_depth,max_depth,d_step))

err_matrix=pd.DataFrame({'max_depth':n,
                         'rmse':rmse_depth},
                        columns=['max_depth','rmse'])
err_matrix
```

```python
print('Minimum rmse is ',round(err_matrix['rmse'].min(),4))
print('The optimal max depth is ',err_matrix['max_depth'][err_matrix['rmse'].idxmin()])
```

```python
'''
Optimize max_leaf_nodes

'''

min_leaf=2
max_leaf=100
l_step=1
rmse_leaf=[]

for i in range(min_leaf,max_leaf,l_step):
    bty_mod=RandomForestRegressor(n_estimators=10, max_depth=5, max_leaf_nodes=i, min_samples_split=2, max_features='sqrt', random_state=42)
    bty_mod.fit(X_train,y_train)
    
    y_pred=bty_mod.predict(X_test)
    rmse_leaf.append(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

time2=time.time()
time_diff=time2-time1

print('total time {}'.format(time_diff))
```

```python
n=list(range(min_leaf,max_leaf,l_step))

err_matrix=pd.DataFrame({'max_leaf':n,
                         'rmse':rmse_leaf},
                        columns=['max_leaf','rmse'])
err_matrix.head(10)
```

```python
print('Minimum rmse is ',round(err_matrix['rmse'].min(),4))
print('The optimal max leaf nodes # is ',err_matrix['max_leaf'][err_matrix['rmse'].idxmin()])
```
