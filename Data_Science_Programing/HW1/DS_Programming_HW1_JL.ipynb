```python
# Take player input of choice from Rock, Scissor and Paper

def player_input():
    '''
    Take player input of choice from Rock, Scissor and Paper
    '''
    player_choice=''
    while player_choice not in {'R','S','P'}:
        player_choice = input('Player please make your choice - Rock(R), Scissors(S) or Paper (P): ')
        player_choice = player_choice.upper()
    return (player_choice)

#Decise who wins

def who_win(player_choice,computer_choice):
    '''
    Take player choice and computer choice to decise who wins
    '''
    if player_choice=='R':
        if computer_choice=='S':
            print('Rock smashes scissors! Player wins!')
            return ('player win')
        elif computer_choice=='P':
            print('Paper covers rock. Computer wins!')
            return ('computer win')
        else:
            print('You both played rock. Its a draw!'.format(computer_choice))
            return ('draw')
    elif  player_choice =='P':
        if computer_choice=='R':
            print('Paper covers rock. Player wins!')
            return ('player win')
        elif computer_choice=='S':
            print('Scissors cut paper. Computer wins!')
            return ('computer win')
        else:
            print('You both played paper. Its a draw!')
            return ('draw')
    else:
        if computer_choice=='P':
            print('Scissors cut paper. Player wins!')
            return ('player win')
        elif computer_choice=='R':
            print('Rock smashes scissors! Computer wins!')
            return ('computer win')
        else:
            print('You both played scissors. Its a draw!')
            return ('draw')
    
```

```python
#Computer made 'trained' decision

from numpy.random import choice
def computer_learning_choice(player_choice_rec):
    '''
    Computer will always choose the R, S, or P that beats what the human player picked most frequently.
    '''
    r_count=player_choice_rec.count('R')
    s_count=player_choice_rec.count('S')
    p_count=player_choice_rec.count('P')
    
    count_rec=[r_count,s_count,p_count]

#if the probability for each choice is equal, choose randomly    
    if r_count==s_count==p_count:
        computer_choice = choice(['R','S','P'])
#if not equal,then there could be the probability of 2 of them are equal, or all of them are different...   
    elif r_count==max(count_rec):
        if r_count==s_count:
            computer_choice='R'
        else:
            computer_choice='P'
    elif s_count==max(count_rec):
        if s_count==p_count:
            computer_choice='S'
        else:
            computer_choice='R'
    else:
        if p_count==r_count:
            computer_choice='P'
        else:
            computer_choice='S'
            
    return(computer_choice)
        

### Alternative algrithm to make 'weighted' ramdon choice based on player choices###
# If computer will consider all entries and form a strategy not to loose, here is an alternative
#     r_count=player_choice_rec.count('R')/len(player_choice_rec)
#     s_count=player_choice_rec.count('S')/len(player_choice_rec)
#     p_count=player_choice_rec.count('P')/len(player_choice_rec)
#     r_weight=(r_count+s_count)/2
#     s_weight=(s_count+p_count)/2
#     p_weight=(p_count+r_count)/2
#     weight=[r_weight,s_weight,p_weight]
#     rsp=['R','S','P']
#     computer_choice = choice(rsp,p=weight)
```

```python
def game():
    player_choice_rec=['R','S','P']
    computer_win_ct=0
    player_win_ct=0
    player_choice=['']
    while True:
        player_choice=player_input()
        computer_choice=computer_learning_choice(player_choice_rec)
        result = who_win(player_choice,computer_choice)
        
        if result=='player win':
            player_win_ct+=1
        if result=='computer win':
            computer_win_ct+=1
        player_choice_rec.append(player_choice)
        game_on=''
        while game_on.upper() not in {'Y','N'}:
            game_on=input('Do you want to play again? Y or N ')
        if game_on.upper()=='N':
            break
    your_win_rate=player_win_ct/(player_win_ct+computer_win_ct)*100
    
    print('You play {} times!\n'.format(len(player_choice_rec)-3))
    print('You win {} times!\n'.format(player_win_ct))
    print('Your win rate is {}%! Good job!'.format(round(your_win_rate,2)))
```

## moving_window

```python
#(a) Convert each tweet into a dictionary of phrases (6 points)


def moving_window(tweet: str):
    """
    divide A tweet into several unique 3-word phrases
    :param tweet: Str type
    :return: A dictionary. Keys are 3-word phrases
    """
    str_list = tweet.split(" ")
    result = {}
    blank = " "
    for i in range(0, len(str_list) - 2):
        add = blank.join(str_list[i:i + 3])
        if add not in result.keys():
            result[add] = 1
    return result


#Codee testing

f = open("Santa.txt", "r")
tweets = f.read()
moving_window(tweets)
```

## cosine

```python
#(b) Calculate similarity between two tweets


def cosine(tweet_1: dict, tweet_2: dict):
    """
    calculate cosine similarity between two tweets
    :param tweet_1: A dictionary. Keys are 3-word phrases of this tweet
    :param tweet_2: A dictionary. Keys are 3-word phrases of this tweet
    :return: cosine similarity between this two tweets
    """
    n_1 = len(tweet_1.keys())
    n_2 = len(tweet_2.keys())
    matches = 0
    for k in tweet_1.keys():
        if k in tweet_2.keys():
            matches += 1
    return float(matches) / math.sqrt(n_1 * n_2)
```

## get tweets and test

```python
#(c) Read in tweets, and output near-duplicates

f = open("Santa.txt", "r")
tweets = f.readlines()
for i in range(0, len(tweets)):
    print(f"tweet {i} : {tweets[i]}")
for i in range(0, len(tweets)):
    for j in range(0, i):
        sim = cosine(moving_window(tweets[i]), moving_window(tweets[j]))
        if sim > 0.5:
            print(f"tweet {i} and tweet {j} is near-duplicate, consine similarity is {sim}")
```

```python
"""The Google of Quotes"""
"""All of the s parameters have a default value of fullText which is the quotes.txt file provided for this class"""
import re
import math
```

```python
fullText = 'quotes.txt'

"""
(a)Build a list of full quotes
takes one parameter with the quotes.txt as a default. 
This function goes through each line and turns it into a list of quotes
"""
def readQuotes(s = fullText):
    newList = []
    i = 0
    fb = open(s,'r')
    quote = fb.readline()
    author = fb.readline()
    while quote != '':
        newList.append(quote.rstrip('\n') + " - " + author.rstrip('\n'))
        quote = fb.readline()
        author = fb.readline()
        i+=1
    return newList
    fb.close()
    
#Code testing

quote_author=readQuotes(fullText)
quote_author
```

```python
"""
(b) Words from a full quotes
Takes one parameter, quote, and splits up the quote into a list of the strings used in the quote
"""
def splitQuote(quote):
    lQuote = quote.lower()
    sQuote = re.split('[^\w]',lQuote)
    while "" in sQuote:
        sQuote.remove('')
    return sQuote

#Code testing

split_quote=splitQuote('I have a hobby...I have the world`s largest collection of sea shells. I keep it scattered on beaches all over the world. Maybe you`ve seen some of it...')
split_quote
```

```python
"""
(c)Postings-list
Takes one parameter, s and reads in the quotes. Then takes in the read in quotes and splits them 
into lists of individual words. Then returns a dictionary with key quote, 
value a dictionary with key each word and value the number of times the word is used in the quote."""

def postListingDict(s = fullText):
    fullDict = {}
    listOfQuotes = readQuotes(s)
    for quote in listOfQuotes:
        
        fullList = splitQuote(quote)
        checkDict = {}
        for item in fullList:
            if(item in checkDict):
                checkDict[item] +=1
            else:
                checkDict[item] = 1
        fullDict[quote] = checkDict
    return fullDict

#Code testing

finalDict=postListingDict(fullText)
finalDict
```

```python
"""
(d)Build the reverse postings-list dictionary
Takes a parameter s and takes dictionary and flips it by rearranging the keys and the values
"""
def reversePostingListDict(s = fullText):
    quoteDict = postListingDict(s)
    reverseDict = {}
    for item, values in quoteDict.items():
        for ourKey in values.keys():
            if ourKey in reverseDict.keys():
                reverseDict[ourKey][item] = values[ourKey]
            else:
                reverseDict[ourKey] = {item:values[ourKey]}
    return reverseDict

#Code testing
reversePostingListDict(fullText)
```

```python
"""
(e)Write a TF-IDF function
Takes choiceword and quote as parameters and counts how many times a given word is said in each quote
it is in and the total times it is in a quote.
Then produces the TFIDF which is the product of TF 
(number of occurrences in the quote/max value for occurrences in the quote)
and the IDF (log(total number of quotes/number of quotes that the given word is in))
"""

    
def TFIDF(choiceWord,quote,s = fullText):
    word = choiceWord.lower()
    reverse = reversePostingListDict(s)
    forward = postListingDict(s)
    if quote not in forward.keys():
        return 0
    elif word not in forward[quote].keys():
        return 0
    else:  
        w = forward[quote][word]
        q = max(forward[quote].values())
    TF = w/q
    N = len(readQuotes(s))
    numQuotes = len(reverse[word].values())
    IDF = math.log(N/numQuotes)
    TFIDF = TF * IDF
    return round(TFIDF,3)

#Code testing
choiceWord='entertainer'
fb = open(fullText, 'r')
text=fb.read()
TFIDF(choiceWord,'An actor is at most a poet and at least an entertainer. - Marlon Brando',)
```

```python
"""
(f) Quote search using a single word
Takes in the choiceWord and s and takes the reversePosting List of the full text. 
Then checks where the word is present and creates a dictionary with the keys being 
each quote the word is in and the values being TFIDF of the word in the given quote.
"""
def singleSearch(choiceWord,s = fullText):
    word = choiceWord.lower()
    reverse = reversePostingListDict(s)
    if word not in reverse.keys():
        return {}
    searchedDict = reverse[word]
    for key in searchedDict.keys():
        searchedDict[key] = TFIDF(word,key,s)
    return searchedDict

#Code testing
singleSearch('happy',)
```

```python
"""Takes a parameter of a list of words and a full text and uses singleSearch for each 
of the words and then merges the dictionary by checking if for repeats and adding the 
values when there is a repeat 
"""
def multipleSearch(listOfWords,s = fullText):
    multiDictionary = singleSearch(listOfWords[0],s)
    if len(listOfWords) == 1:
        return 
    words = listOfWords[1:]
    for item in words:
        item = item.lower()
        temp = singleSearch(item,s)
        for key in temp.keys():
            if key in multiDictionary.keys():
                multiDictionary[key] = multiDictionary[key] + TFIDF(item,key,s)
            else:
                multiDictionary[key] = TFIDF(item,key,s)
    return multiDictionary

#Code testing
multipleSearch(['heart','mind','disease'])
```
